```python
import numpy as np
import pandas as pd

# Function to load data from CSV file
def load_data(input_file):
    return pd.read_csv(input_file)

# Function to add Laplace noise to numerical data
def laplace_mechanism(df, epsilon=1.0, sensitivity=1.0):
    noisy_df = df.copy()
    for column in df.select_dtypes(include=[np.number]).columns:  # Only apply Laplace noise to numerical columns
        # Compute Laplace noise based on sensitivity and epsilon
        scale = sensitivity / epsilon
        noise = np.random.laplace(0, scale, size=df[column].shape)
        noisy_df[column] = df[column] + noise
    return noisy_df

# Function to mask sensitive data (like phone numbers)
def mask_data(df):
    masked_df = df.copy()
    if 'phone' in df.columns:
        masked_df['phone'] = masked_df['phone'].apply(lambda x: 'XXX-XXX-' + x[-4:])
    return masked_df

# Function to apply k-anonymity (generalization of age and zip code)
def k_anonymity(df, k=2):
    anonymized_df = df.copy()
    
    # Generalize age into ranges
    if 'age' in df.columns:
        anonymized_df['age'] = anonymized_df['age'].apply(lambda x: f"{(x // 10) * 10}-{(x // 10) * 10 + 9}")
    
    # Generalize zip code (e.g., only show the first 3 digits)
    if 'zip_code' in df.columns:
        anonymized_df['zip_code'] = anonymized_df['zip_code'].apply(lambda x: x[:3] + "XX")
    
    return anonymized_df

# Main function to apply perturbation techniques and save the result to a new file
def apply_perturbation(input_file, output_file, epsilon=1.0, sensitivity=1.0):
    # Load input data
    data = load_data(input_file)
    
    # Apply Laplace mechanism to add noise
    noisy_data = laplace_mechanism(data, epsilon=epsilon, sensitivity=sensitivity)
    
    # Apply masking for sensitive data (phone numbers)
    masked_data = mask_data(noisy_data)
    
    # Apply k-anonymity (generalize age and zip code)
    anonymized_data = k_anonymity(masked_data)
    
    # Save the resulting data to a new CSV file
    anonymized_data.to_csv(output_file, index=False)
    print(f"Anonymized data saved to {output_file}")

# Usage
input_file = 'input_data.csv'  # Path to your input CSV file
output_file = 'output_data.csv'  # Path to save the output CSV file
apply_perturbation(input_file, output_file, epsilon=1.0, sensitivity=1.0)












2nd option :





```
